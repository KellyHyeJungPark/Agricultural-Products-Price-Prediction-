```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a= pd.read_csv('../data/Agri_all.csv')
```

```python
garlic = a[a['Product'] == '마늘']
potato = a[a['Product'] == '감자']
s_potato = a[a['Product'] == '고구마']
leaf = a[a['Product'] == '깻잎']
```

```python
garlic['5mv'] = garlic['Price'].rolling(window=5).mean()
garlic['30mv'] = garlic['Price'].rolling(window=30).mean()
garlic['60mv'] = garlic['Price'].rolling(window=60).mean()
garlic['90mv'] = garlic['Price'].rolling(window=90).mean()
garlic['120mv'] = garlic['Price'].rolling(window=120).mean()

potato['5mv'] = potato['Price'].rolling(window=5).mean()
potato['30mv'] = potato['Price'].rolling(window=30).mean()
potato['60mv'] = potato['Price'].rolling(window=60).mean()
potato['90mv'] = potato['Price'].rolling(window=90).mean()
potato['120mv'] = potato['Price'].rolling(window=120).mean()

s_potato['5mv'] = s_potato['Price'].rolling(window=5).mean()
s_potato['30mv'] = s_potato['Price'].rolling(window=30).mean()
s_potato['60mv'] = s_potato['Price'].rolling(window=60).mean()
s_potato['90mv'] = s_potato['Price'].rolling(window=90).mean()
s_potato['120mv'] = s_potato['Price'].rolling(window=120).mean()

leaf['5mv'] = leaf['Price'].rolling(window=5).mean()
leaf['30mv'] = leaf['Price'].rolling(window=30).mean()
leaf['60mv'] = leaf['Price'].rolling(window=60).mean()
leaf['90mv'] = leaf['Price'].rolling(window=90).mean()
leaf['120mv'] = leaf['Price'].rolling(window=120).mean()
```

```python
garlic.to_csv('garlic.csv')
potato.to_csv('potato.csv')
s_potato.to_csv('s_potato.csv')
leaf.to_csv('leaf.csv')
```
