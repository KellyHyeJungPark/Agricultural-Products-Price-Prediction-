```python
df2 = pd.read_excel('C:\\Users\\Woo\\Documents\\GitHub\\Agricultural-Products-Price-Prediction-\\Data Collection\\data\\생산자물가지수_품목별_연도별_월별.xlsx')
df2 = df2.rename(columns={'계정코드별':'YM'})
df2["YM"] = df2["YM"].map(lambda x : x.split()[0])
df2 = df2.replace("-", np.NaN)
df2 = df2.dropna(axis=0)
df2 = df2[df2["YM"].isin(crop_list)]
for i in df2["YM"]:
    df2["YM"] = df2["YM"].replace(i,f"생산자물가지수_{i}")
df2
```
