```python
df2 = pd.read_excel('/content/drive/Shareddrives/Agricultural Products Price Prediction/data/생산자물가지수_품목별_연도별_월별.xlsx')
df2 = df2.rename(columns={'계정코드별':'YM'})
df2["YM"] = df2["YM"].map(lambda x : x.split()[0])
df2 = df2.replace("-", np.NaN)
df2 = df2.dropna(axis=0)
for i in df2["YM"]:
    df2["YM"] = df2["YM"].replace(i,f"생산자물가지수_{i}")
df2.T
```

```python
df3 = pd.read_csv('/content/drive/Shareddrives/Agricultural Products Price Prediction/data/소비자물가지수_품목별_연도별_월별_filtered.csv')
df3 = df3.replace("-", np.NaN)
df3 = df3.rename(columns={'품목별':'YM'})
df3 = df3.dropna(axis="rows")
df3 = df3.drop(index=[0,1,2,10,38])
for i in df3["YM"]:
    df3["YM"] = df3["YM"].replace(i,f"생산자물가지수_{i}")
df3.T
```
