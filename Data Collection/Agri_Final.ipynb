```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib
%config InlineBackend.figure_format = 'retina'

root_path = "../"
data_path = f"{root_path}data"
```

```python
market = pd.read_csv(f'{data_path}/AgriMarket.csv')
weather = pd.read_csv(f'{data_path}/AgriWeather.csv')
market.shape, weather.shape
```

```python
market['YMD'] = market['YMD'].astype('str')
```

```python
market['YMD'] = pd.to_datetime(market['YMD'])
weather['YMD'] = pd.to_datetime(weather['YMD'])
```

```python
market
```

```python
potato_m = market[market['Product'] == '감자']
potato_w = weather[weather['Crop_Name'] == '감자']
potato = pd.merge(potato_m, potato_w, on='YMD', how='left')
potato = potato.reset_index()
potato = potato.drop(columns='index')
potato
```

```python
garlic_m = market[market['Product'] == '마늘']
galic_w = weather[weather['Crop_Name'] == '마늘']
garlic = pd.merge(garlic_m, galic_w, on='YMD', how='left')
garlic = garlic.reset_index()
garlic = garlic.drop(columns='index')
garlic
```

```python
s_potato_m = market[market['Product'] == '고구마']
s_potato_w = weather[weather['Crop_Name'] == '고구마']
s_potato = pd.merge(s_potato_m, s_potato_w, on='YMD', how='left')
s_potato = s_potato.reset_index()
s_potato = s_potato.drop(columns='index')
s_potato
```

```python
Sesame_leaf_m = market[market['Product'] == '깻잎']
Sesame_leaf_w = weather[weather['Crop_Name'] == '깻잎']
Sesame_leaf = pd.merge(Sesame_leaf_m, Sesame_leaf_w, on='YMD', how='left')
Sesame_leaf = Sesame_leaf.reset_index()
Sesame_leaf = Sesame_leaf.drop(columns='index')
Sesame_leaf
```

```python
garlic = garlic.drop(columns='Crop_Name')
Sesame_leaf = Sesame_leaf.drop(columns='Crop_Name')
potato = potato.drop(columns='Crop_Name')
s_potato = s_potato.drop(columns='Crop_Name')
```

```python
garlic = garlic.fillna(method = 'bfill')
Sesame_leaf = Sesame_leaf.fillna(method = 'bfill')
potato = potato.fillna(method = 'bfill')
s_potato = s_potato.fillna(method = 'bfill')
```
